```python
import pandas as pd
import geopandas as gpd
import os
os.getcwd()
```

```python
demand_df = pd.read_csv('../data/processed/demand_surface.csv')
facilities_df = pd.read_csv('../data/processed/facilities_with_planning_area.csv')
gdf = gpd.read_file('../data/raw/DataMasterPlan2019PlngAreaBoundaryNoSea.geojson')
```

```python
# Inspect facilities
print("FACILITIES")
print(facilities_df.shape)
print(facilities_df.dtypes)
facilities_df.head()
```

```python
# Inspect GeoJSON
print("GEOJSON")
print(gdf.shape)
print(gdf.dtypes)
gdf.head()
```

```python
print(demand_df['PLN_AREA_N'].sort_values().unique())
print(facilities_df['PLN_AREA_N'].sort_values().unique())
print(gdf['PLN_AREA_N'].sort_values().unique())
```

```python
demand_set = set(demand_df['PLN_AREA_N'])
facilities_set = set(facilities_df['PLN_AREA_N'])
geo_set = set(gdf['PLN_AREA_N'])

print("In facilities but not in demand:", facilities_set - demand_set)
print("In GeoJSON but not in demand:", geo_set - demand_set)
print("In demand but not in GeoJSON:", demand_set - geo_set)
```

```python
facility_counts = facilities_df.groupby('PLN_AREA_N')['facility_id'].agg('count').reset_index()
print(facility_counts.shape)
facility_counts.head()
```

```python
merged_df = demand_df.merge(facility_counts, on='PLN_AREA_N', how='left')
merged_df = merged_df.rename(columns={'facility_id': 'facility_count'})
merged_df['facility_count'] = merged_df['facility_count'].fillna(0).astype(int)

print(merged_df.shape)
merged_df.head(10)
```

## Design Note: Aggregated vs Facility-Level Master Table

We are deliberately working at planning area level for this notebook.
The 148-row facility-level source (`facilities_with_planning_area.csv`) 
is preserved as the row-level source of truth for any future drill-down, 
ABM inputs, or stress testing.

A facility-level master table (one row per facility, all planning area 
context columns repeated) is a valid alternative architecture and would 
simplify certain future queries. That refactor is deferred — the current 
pipeline serves the choropleth dashboard output cleanly.

```python
merged_df = merged_df.merge(gdf[['PLN_AREA_N', 'geometry']], on='PLN_AREA_N', how='left')

merged_df.head()
```

```python
merged_df['pwd_per_facility'] = (merged_df['pwd_estimate'] / merged_df['facility_count']).where(merged_df['facility_count'] > 0)

merged_df
```

```python
merged_df['no_facility'] = merged_df['facility_count'] == 0
merged_df.head()
```

```python
#average pwd per facility 

pwd_per_facility_sort = merged_df[merged_df['facility_count'] > 0][['PLN_AREA_N', 'pwd_estimate', 'facility_count', 'pwd_per_facility']].sort_values('pwd_per_facility', ascending=False)
pwd_per_facility_sort
```

```python
#Population with no facility. undersupply list

dd_no_ss_sort = merged_df[
    (merged_df['no_facility'] == True) & 
    (merged_df['pwd_estimate'] > 0)
][['PLN_AREA_N', 'pwd_estimate', 'facility_count', 'pwd_per_facility']].sort_values('pwd_estimate', ascending=False)

dd_no_ss_sort
```

```python
#No population, No facilities

No_pop_no_fac = merged_df[
    (merged_df['no_facility'] == True) & 
    (merged_df['pwd_estimate'] == 0)
][['PLN_AREA_N', 'pwd_estimate', 'facility_count', 'pwd_per_facility']]


No_pop_no_fac
```

```python
#No Population, with facilities

No_pop_withfac = merged_df[
    (merged_df['no_facility'] == False) & 
    (merged_df['pwd_estimate'] == 0)
][['PLN_AREA_N', 'pwd_estimate', 'facility_count', 'pwd_per_facility']]


No_pop_withfac
```

```python
#pwd largest demand

merged_df[
    (merged_df['no_facility'] == False) & 
    (merged_df['pwd_estimate'] > 0)
][['PLN_AREA_N', 'pwd_estimate', 'facility_count', 'pwd_per_facility']].sort_values('pwd_estimate', ascending=False)
```

```python
# Option 1: mean of ratios
avg_pwd_per_facility = pwd_per_facility_sort['pwd_per_facility'].mean()
print(f"Mean pwd_per_facility across served areas: {avg_pwd_per_facility:.1f}")
```

```python
total_pwds = pwd_per_facility_sort['pwd_estimate'].sum()
total_facilities = pwd_per_facility_sort['facility_count'].sum()
weighted_avg = total_pwds / total_facilities

print(f"Total PWDs in served areas: {total_pwds}")
print(f"Total facilities: {total_facilities}")
print(f"Weighted avg PWDs per facility: {weighted_avg:.1f}")
```

```python
total_unserved_pwds = dd_no_ss_sort['pwd_estimate'].sum()
print(f"Total PWDs in planning areas with no facility: {total_unserved_pwds}")
```

```python
from openpyxl import Workbook

# Drop geometry for Excel - not compatible
merged_excel = merged_df.drop(columns=['geometry'])

# Write to Excel with multiple sheets
with pd.ExcelWriter('../data/processed/dementia_analysis.xlsx', engine='openpyxl') as writer:
    merged_excel.to_excel(writer, sheet_name='Full Analysis', index=False)
    pwd_per_facility_sort.to_excel(writer, sheet_name='Served Areas', index=False)
    dd_no_ss_sort.to_excel(writer, sheet_name='Undersupply', index=False)
    No_pop_no_fac.to_excel(writer, sheet_name='Noise', index=False)

print("Export complete")
```
